## Random distributions

### Standard Normal distribution with $\mu$ = 0 and $\sigma$ = 1 

```python
x = np.random.normal(size=10000)
mu = np.mean(x) #MEAN
sigma = np.std(x) #STANDARD DEVIATION
print(f"length of x = {len(x)}")
counts, bins = np.histogram(x, bins=50)
plt.hist(bins[:-1], bins, weights=counts)
plt.xlabel("bins")
plt.ylabel("frequency")
plt.text(1.5,500,"Normal distribution")
plt.text(1.5,450,f"mu = {round(mu,2)}")
plt.text(1.5,400,f"sigma = {round(sigma,2)}")
```

### Displaced gaussian, i.e., a different $\mu$

```python
y = np.random.normal(loc=50, scale=1, size=10000)
print(f"length of y = {len(y)}")
mu = np.mean(y) #MEAN
sigma = np.std(y) #STANDARD DEVIATION
counts, bins = np.histogram(y, bins=50)
plt.hist(bins[:-1], bins, weights=counts)
plt.xlabel("bins")
plt.ylabel("frequency")
plt.text(51,500,"Normal distribution")
plt.text(52,450,f"mu = {round(mu,2)}")
plt.text(52,400,f"sigma = {round(sigma,2)}")
```

### Gaussian with an arbitrary $\mu$ and $\sigma$ 

```python
z = np.random.normal(loc=50, scale=10, size=10000)
print(f"length of z = {len(z)}")
mu = np.mean(z) #MEAN
sigma = np.std(z) #STANDARD DEVIATION
counts, bins = np.histogram(z, bins=50)
plt.hist(bins[:-1], bins, weights=counts)
plt.xlabel("bins")
plt.ylabel("frequency")
plt.text(68,500,"Normal distribution")
plt.text(68,450,f"mu = {round(mu,2)}")
plt.text(68,400,f"sigma = {round(sigma,2)}")
```

## Seeding

In order to ensure that our code provides exactly the same results
each time it is run, we can set a *random seed* 
using the 
`np.random.default_rng()` function.
This function takes an arbitrary, user-specified integer argument. If we set a random seed before 
generating random data, then re-running our code will yield the same results. The
object `rng` has essentially all the random number generating methods found in `np.random`. Hence, to
generate normal data we use `rng.normal()`.

```python
rng = np.random.default_rng(202)
rng.normal(loc=50, scale=10, size=10)
```

```python
a = rng.standard_normal((2,3))
a
```

```python
a.var(axis=0)
```

```python
a.mean(axis=1)
```

## Figures, Subfigures and Scatter Plots

### pyplot.plot

```python
fig, ax = plt.subplots(figsize=(6,6))
x = rng.standard_normal(1000)
y = rng.standard_normal(1000)
ax.plot(x,y);
```

### Scatter plots

```python
fig, ax = plt.subplots(figsize=(6,6))
x = rng.standard_normal(1000)
y = rng.standard_normal(1000)
ax.plot(x,y,'o');
```

```python
fig, ax = plt.subplots(figsize=(6,6))
x = rng.standard_normal(1000)
y = rng.standard_normal(1000)
ax.scatter(x,y,marker='o');
```

### Adjusting Size

```python
fig, ax = plt.subplots(figsize=(6,6))
x = rng.standard_normal(1000)
y = rng.standard_normal(1000)
ax.scatter(x,y,marker='o');
fig.set_size_inches(12,3)
```

### Grids

```python
fig, axes = plt.subplots(nrows=2, ncols=3 ,figsize=(16,8))
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3 ,figsize=(16,8),sharex=True,sharey=True)
markers = np.array([['o','+','*'],['.','+',"o"]])
for ax_row in range(len(axes[:,0])):
    for ax_col in range(len(axes[0,:])) :
        axes[ax_row,ax_col].scatter(x,y,marker=markers[ax_row,ax_col])
```

### Contour plots

Countour plots are 2D representation of 3D data.
x and y should be 1D and z should be a 2D array.

```python
z = np.multiply.outer([1,2,3,4],[4,3,2,1]) #creates a multiplication table which is a 2D array
z
```

```python
fig , ax = plt.subplots(figsize=(8,8))
ax.contour([1,2,3,4],[4,3,2,1],z)
```

### Heat Map

```python
fig , ax = plt.subplots(figsize=(8,8))
ax.imshow(z)
```

## Indexing and Slicing

```python
A = np.arange(1,10,1).reshape(3,3)
A
```

```python
A[1,2] #2nd row and 3rd column
```

```python
A[[0,2]] #choose 1st and 3rd row
```

```python
A[:,[0,1]] #choose 1st and 2nd Column
```

```python
A[[0,1],[0,1]] # chooses the  diagonal instead of the submatrix!
```

```python
A[[0,1]][:,[0,1]] #complex way to get the submatrix
```

```python
np.ix_([3,4,5],[4,5,6]) # create a mesh
```

```python
idx = np.ix_([0,1],[0,1]) # more efficient way to get submatrix
A[idx]
```
